Danceability = Describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.

Energy = Measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy.

Duration_ms: duration of track in milliseconds

Key = The estimated overall key of the track. Integers map to pitches using standard Pitch Class notation. E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on. If no key was detected, the value is set to -1.

Loudness = The overall loudness of a track in decibels (dB). Values typical range between -60 and 0 db. A positive dB means that the sound is a few times louder than the threshold, while a negative dB means that you are a few times softer than that threshold.

Instrumentalness = Measure from 0.0 to 1.0 and represents the amount of vocals in the song. The closer it is to 1.0, the more instrumental the song is.

Speechiness = Measures from 0.0 to 1.0 and detects the presence of spoken words in a track. If the speechiness of a song is above 0.66, it is probably made of spoken words, a score between 0.33 and 0.66 is a song that may contain both music and words, and a score below 0.33 means the song does not have any speech.

Liveness = Likelihood measure from 0.0 to 1.0 and indicates the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live.

Valence = A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track.

Tempo = The overall estimated tempo of a track in beats per minute (BPM).

Time_signature = An estimated overall time signature of a track. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure).

```python
# Danceability
# Energy
# Key
# Loudness
# Instrumentalness
# Popularity          

figure,axesbi=plt.subplots(3,2,figsize=(30,40))

a=df.groupby('track_genre')['danceability'].mean().sort_values().head(20).plot(kind='barh',ax=axesbi[0,0])
a.set_title('danceability',fontsize=25)
for label in (a.get_xticklabels() + a.get_yticklabels()):
    label.set_fontsize(20)


b=df.groupby('track_genre')['energy'].mean().sort_values().head(20).plot(kind='barh',ax=axesbi[0,1],color='orange')
b.set_title('energy',fontsize=25)
for label in (b.get_xticklabels() + b.get_yticklabels()):
    label.set_fontsize(20)


c=df.groupby('track_genre')['key'].mean().sort_values().head(20).plot(kind='barh',ax=axesbi[1,0],color='green')
c.set_title('key',fontsize=25)
for label in (c.get_xticklabels() + c.get_yticklabels()):
    label.set_fontsize(20)


d=df.groupby('track_genre')['loudness'].mean().sort_values().head(20).plot(kind='barh',ax=axesbi[1,1],color='yellow')
d.set_title('loudness',fontsize=25)
for label in (d.get_xticklabels() + d.get_yticklabels()):
    label.set_fontsize(20)


e=df.groupby('track_genre')['instrumentalness'].mean().sort_values().head(20).plot(kind='barh',ax=axesbi[2,0],color='pink')
e.set_title('instrumentalness',fontsize=25)
for label in (e.get_xticklabels() + e.get_yticklabels()):
    label.set_fontsize(20)
    

f=df.groupby('track_genre')['popularity'].mean().sort_values().head(20).plot(kind='barh',ax=axesbi[2,1],color='red')
f.set_title('popularity',fontsize=25)
for label in (f.get_xticklabels() + f.get_yticklabels()):
    label.set_fontsize(20)    
```

```python
Genre=df['track_genre'].value_counts(sort=True).tail(10)
Genre
```

```python
# Bar plot with respect to Genres:

fig, ax = plt.subplots(figsize = (12, 10))

ax = sns.barplot(x = Genre.values, y = Genre.index,  orient = 'h', edgecolor = 'black', ax = ax)

ax.set_xlabel('Counts', c ='black', fontsize = 12)
ax.set_ylabel('Genre', c = 'black', fontsize = 12)
ax.set_title('Top Most Popular Genre in Dataset', c = 'black', fontsize = 14, weight = 'bold')


for i in ax.containers:
    ax.bar_label(i,)

plt.show()
```

```python
# Creating most popular words with respect to genres:

plt.figure(figsize = (20, 14))

def visualize_word_counts(counts):
    wc = WordCloud(max_font_size=130, min_font_size=25, colormap='tab20', background_color='white', prefer_horizontal=.95, width=2100, height=700, random_state=0)
    cloud = wc.generate_from_frequencies(counts)
    plt.figure(figsize=(18,15))
    plt.imshow(cloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
    
visualize_word_counts(Genre)
```

Based on the word count visualization we can say that "romance" have lower number of counts as compared to other genres.
